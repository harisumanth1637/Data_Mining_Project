Splitting the data

```python
# Sepratating & assigning features and target columns to X & y
Y = df['Phishing']
X = df.drop('Phishing',axis=1)
X.shape, Y.shape
     
```

```python
# Splitting the dataset into train and test sets: 80-20 split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y,test_size = 0.2, random_state = 12)
X_train.shape, X_test.shape
     
```

Machine Learning Models
The supervised machine learning models (classification) 
For phishing website detection, several machine learning models can be used effectively. Here are some commonly used models for binary classification tasks like phishing detection:

Logistic Regression: A simple and efficient linear model that works well for binary classification tasks.
Decision Tree: A tree-based model that can capture non-linear relationships and interactions between features.
Random Forest: An ensemble of decision trees that improves performance by averaging the predictions of multiple trees.
Gradient Boosting: Boosting algorithm that builds trees sequentially, with each tree correcting the errors of the previous one, leading to high accuracy.
Support Vector Machine (SVM): Effective for high-dimensional data, SVM finds the hyperplane that best separates the classes in feature space.
Neural Networks: Deep learning models, such as Multilayer Perceptrons (MLPs), can capture complex patterns in the data but require more data and computational resources.
AdaBoost: Another boosting algorithm that combines multiple weak learners to create a strong classifier, often effective for binary classification tasks.

```python
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn import svm, tree, ensemble, naive_bayes, neural_network, neighbors
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# Create a dictionary of models
models = {
    'SVM': svm.LinearSVC(),
    'Random Forest': ensemble.RandomForestClassifier(n_estimators=60),
    'Decision Tree': tree.DecisionTreeClassifier(),
    'AdaBoost': ensemble.AdaBoostClassifier(),
    'Gaussian Naive Bayes': naive_bayes.GaussianNB(),
    'Neural Network': neural_network.MLPClassifier(alpha=1),
    'K-Nearest Neighbors': neighbors.KNeighborsClassifier()
}

# Initialize results dictionary
results = {'Model': [], 'Accuracy (Train)': [], 'Precision (Train)': [], 'Recall (Train)': [],
           'F1-score (Train)': [], 'Accuracy (Test)': [], 'Precision (Test)': [], 'Recall (Test)': [],
           'F1-score (Test)': []}

# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

for name, model in models.items():
    # Train the model
    model.fit(X_train_scaled, y_train)
    
    # Predictions on train and test set
    y_train_pred = model.predict(X_train_scaled)
    y_test_pred = model.predict(X_test_scaled)
    
    # Calculate evaluation metrics
    train_accuracy = accuracy_score(y_train, y_train_pred)
    test_accuracy = accuracy_score(y_test, y_test_pred)
    precision_train = precision_score(y_train, y_train_pred)
    precision_test = precision_score(y_test, y_test_pred)
    recall_train = recall_score(y_train, y_train_pred)
    recall_test = recall_score(y_test, y_test_pred)
    f1_train = f1_score(y_train, y_train_pred)
    f1_test = f1_score(y_test, y_test_pred)
    
    # Append results to dictionary
    results['Model'].append(name)
    results['Accuracy (Train)'].append(train_accuracy)
    results['Precision (Train)'].append(precision_train)
    results['Recall (Train)'].append(recall_train)
    results['F1-score (Train)'].append(f1_train)
    results['Accuracy (Test)'].append(test_accuracy)
    results['Precision (Test)'].append(precision_test)
    results['Recall (Test)'].append(recall_test)
    results['F1-score (Test)'].append(f1_test)

# Convert results to DataFrame
results_df = pd.DataFrame(results)

# Plotting Train vs Test performance
plt.figure(figsize=(12, 8))
plt.plot(results_df['Model'], results_df['Accuracy (Train)'], marker='o', label='Train Accuracy')
plt.plot(results_df['Model'], results_df['Accuracy (Test)'], marker='o', label='Test Accuracy')
plt.title('Model Performance (Train vs Test)')
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Display results
print(results_df)
```

```python
# Calculate mean performance metrics for all models
mean_results = results_df.mean()

# Determine the best performing model based on a chosen metric (e.g., accuracy)
best_model = results_df.loc[results_df['Accuracy (Test)'].idxmax()]

# Display the summary of performance metrics
print("\nMean Performance Metrics Across Models:")
print(mean_results)

# Display the best performing model
print("\nBest Performing Model (Based on Test Accuracy):")
print(best_model)
```
